```python
# import libraries
import googletrans
from googletrans import Translator
import pandas as pd
```

# Import data

```python
#Retrive world cities list
data = pd.read_excel('/content/drive/MyDrive/worldcities.xlsx')
italian_cities = data[data['country'] == 'Italy']
```

# Translate

```python
translator = Translator()
def translation(word):
  t_word = translator.translate(word, src='en', dest='it')
  return t_word.text
```

```python
italian_cities['city_ita'] = italian_cities['city'].apply(translation)
italian_cities['region_ita'] = italian_cities['admin_name'].apply(translation)
italian_cities
```

```python
ita_city = italian_cities[['city_ita','region_ita']].copy()
ita_city
```

# Check

```python
# Check unique values
unique = ita_city['region_ita'].unique()
unique
```

```python
# Manually correct bad translations if needed
original_words = ['Campano','veneto']
replace_words = ['Campania','Veneto']
column_to_replace = 'region_ita'

for index,word in enumerate(original_words):
  ita_city[column_to_replace] = ita_city[column_to_replace].replace(word,replace_words[index])

ita_city
```

# Export

```python
excel_file_path = '/content/drive/MyDrive/ItalianCities.xlsx'
ita_city.to_excel(excel_file_path,'Italian cities and regions')
print(f"DataFrame has been exported to {excel_file_path}")
```
